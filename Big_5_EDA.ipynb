```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Final_df = pd.read_csv('Final.csv')
```

```python
# Check if the specified columns are present in the Final_dfset
if 'xG' in Final_df.columns and 'market_value_in_eur' in Final_df.columns:
    # Plotting the scatter plot for market value vs xG
    plt.figure(figsize=(10, 6))
    plt.scatter(Final_df['xG'], Final_df['market_value_in_eur'], alpha=0.6)
    plt.title("Market Value vs Expected Goals (xG)")
    plt.xlabel("Expected Goals (xG)")
    plt.ylabel("Market Value in EUR")
    plt.grid(True)
    plt.show()
else:
    missing_columns = [col for col in ['xG', 'market_value_in_eur'] if col not in Final_df.columns]
    print(f"Columns missing in the Final_df: {missing_columns}")
```

```python
# Check if the specified columns are present in the Final_dfset
if 'xA' in Final_df.columns and 'market_value_in_eur' in Final_df.columns:
    # Plotting the scatter plot for market value vs xG
    plt.figure(figsize=(10, 6))
    plt.scatter(Final_df['xA'], Final_df['market_value_in_eur'], alpha=0.6)
    plt.title("Market Value vs Expected Assists (xA)")
    plt.xlabel("Expected Assists (xA)")
    plt.ylabel("Market Value in EUR")
    plt.grid(True)
    plt.show()
else:
    missing_columns = [col for col in ['xA', 'market_value_in_eur'] if col not in Final_df.columns]
    print(f"Columns missing in the Final_df: {missing_columns}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check if the specified columns are present in the dataset
if 'Age' in Final_df.columns and 'market_value_in_eur' in Final_df.columns:
    plt.figure(figsize=(10, 4))
    sns.kdeplot(data =Final_df, x='Age', y='market_value_in_eur', fill=True, cmap="viridis")
    plt.title("Distribution Curve of Market Value by Age")
    plt.xlabel("Age")
    plt.ylabel("Market Value in EUR")
    plt.grid(True)
    plt.show()
else:
    missing_columns = [col for col in ['Age', 'market_value_in_eur'] if col not in Final_df.columns]
    print(f"Columns missing in the data: {missing_columns}")
```

```python
# Grouping the data by club ('Squad') and summing the market value for each club
club_market_value = Final_df.groupby('Squad')['market_value_in_eur'].sum().sort_values()

# Plotting the bar chart
plt.figure(figsize=(12, 14))
club_market_value.plot(kind='barh')
plt.title("Total Market Value by Club")
plt.xlabel("Total Market Value in EUR")
plt.ylabel("Club")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
